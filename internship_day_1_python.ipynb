q1

```python

def count_vowels(text):
    count = 0
    for letter in text:
        if letter in "aeiouAEIOU":
            count += 1
    return count
```

q2

```python


def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

```

q3

```python

def count_character(text, character):
    count = 0
    for letter in text:
        if letter == character:
            count += 1
    return count
```

q4

```python


def count_odd_digits(num):
    count = 0
    for digit in str(num):
        if int(digit) % 2 != 0:
            count += 1
    return count

```

```python


def remove_spaces(text):
    result = ""
    for letter in text:
        if letter != " ":
            result += letter
    return result


```
